# line plots with multifacets


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('dots')
df.head()
```

```python
# defining color palette
palette = sns.color_palette('rocket_r')
df.columns
```

```python
# plot lineplot
sns.relplot(
    data=df,
    x="time",
    y="firing_rate",
    hue="coherence",
    size="choice",
    col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=0.75, facet_kws=dict(sharex=False)
)
plt.show()
```
