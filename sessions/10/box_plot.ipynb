# Boxplot in python


```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean, median
import pandas as pd
import numpy as np

sns.set_style("whitegrid")

data = sns.load_dataset('titanic')
```

```python
data.sample(5)
```

```python
sns.boxplot(x="class", y="fare", data=data)
plt.show()
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data.sample(5)
```

```python
sns.boxplot(x="day", y="tip", data=tips_data)
plt.show()
```

```python
sns.barplot(x="day", y="tip", data=tips_data)
plt.show()
```

```python
sns.boxplot(x="day", y="tip", data=tips_data, saturation=0.8)
plt.show()
```

```python
sns.boxplot(x="day", y="tip", data=tips_data)
```

```python
tips_data.describe()
```

```python
sns.boxplot(x=tips_data['total_bill'], y=tips_data['tip'])
plt.show()
```

```python
sns.boxplot(x=tips_data['day'], y=tips_data['tip'])
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
sns.boxplot(x="tip", y="day", data=tips_data)
```

```python
sns.boxplot(x="tip", y="day", data=tips_data, hue="smoker")
```

```python
sns.boxplot(x="tip", y="day", data=tips_data, hue="smoker", palette="Set2", dodge=True)
```

```python
sns.boxplot(x="tip", y="day", data=tips_data, hue="smoker", palette="Set2", dodge=False)
```

```python
# orient = h or v, depending on the data on x and y side.
# Vertical orientation requires numeric `y` variable.
# sns.boxplot(x="tip", y="day", data=tips_data, hue="smoker", palette="Set2", dodge=False, orient="v")

sns.boxplot(x="tip", y="day", data=tips_data, hue="smoker", palette="Set2", dodge=False, orient="h")
```

```python
sns.boxplot(x="survived", y="age", data=data)
plt.show()
```

```python
# FontDict
# font = {'family': 'serif',
#         'color':  'darkred',
#         'weight': 'normal',
#         'size': 16,
#         }
sns.boxplot(x="survived", y="age", data=data)
# plt.xlabel("How many survived", size=12, weight="bold", family="Arial, Helvetica, sans-serif")
plt.xlabel("How many survived", size=12, weight="bold", family="Arial")
plt.ylabel("Age (Years)", size=12, color="red", weight="bold", family="Arial, Helvetica, sans-serif")
plt.title("Box plot of survived and dead from Titanic", size=12, color="green", weight="bold")
plt.show()
```

```python
meanpointprops = {'marker':'+', "markeredgecolor": 'black', "markerfacecolor": 'firebrick'}
sns.boxplot(x="survived", y="age", data=data, showmeans=True, meanprops=meanpointprops)
plt.show()
```
