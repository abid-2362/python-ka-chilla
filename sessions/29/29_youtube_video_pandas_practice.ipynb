```python
import pandas as pd

df = pd.read_csv("../../data/pokemon_data.csv")

# if reading .txt file in csv, specify the delimiter to separate columns of data
# df = pd.read_csv("../../data/pokemon_data.txt", delimiter="\t")

df.head(5)
```

```python
# Getting columns
df.columns
```

```python
# Converting columns into list
list(df.columns)
```

```python
# Read a specific column
# df.Name # If column name consists of 1 word, without special chars. this can be used as well
# OR This can work if there is a space in column name
df["Name"]
```

```python
# Read Each Row
df.iloc[0:]
```

```python
# Read row ranges
df.iloc[0:4]
```

```python
# Read a specific location (Row, Column)
df.iloc[2, 1]
```

```python
# Iterate through rows one by one
for index, row in df.iterrows():
    print("-----\n", "index:", index, "\n", "row:\n", row, "\n----\n\n")
```

```python
# Iterate through rows one by one and read Specific column value
for index, row in df.iterrows():
    print(index, row["Name"], " | ", row["Type 1"])
```

```python
# Getting data in a specific location

# Can be used to filter data with conditions
# Gets data only where Type1 equals to "Fire"
df.loc[df["Type 1"] == "Fire"]
```

```python
# Can be used to filter data with conditions
# Gets data only where Type1 equals to "Grass"
df.loc[df["Type 1"] == "Grass"]
```

```python
df.describe()
```

```python
# Sorting dataframe depending on values
df.sort_values(by="Name", ascending=True)  # ascending = True Or False; False means descending
```

```python
# Sorting dataframe by multiple columns
df.sort_values(by=["Type 1", "HP"])  # ascending = True Or False; False means descending
```

```python
# Sorting dataframe by multiple columns in ascending or descending order
df.sort_values(by=["Type 1", "HP"], ascending=True)  # ascending = True Or False; False means descending
```

```python
# Sorting dataframe by multiple columns in one ascending and 2nd descending
# This means, Type 1 will be ascending and HP will be descending
df.sort_values(by=["Type 1", "HP"], ascending=[1, 0])  # [1= ascending, 0=descending]
```

## Making changes the data

```python
# Adding new column with total of all HP, Attack, Defense, Sp. Atk, Sp. Defense, Speed
df['Total'] = df['HP'] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
df
```

```python
# Dropping some columns
df = df.drop(columns=["Total"])

df
```

```python
# adding column by calculating values dynamically
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head()
```

```python
cols = list(df.columns.values)
cols
```

```python
cols[0:4]
```

```python
cols[-1]
```

```python
[cols[-1]]
```

```python
# Rearranging columns in different ways
# TypeError: can only concatenate list (not "str") to list
# df = df[cols[0:4] + cols[-1] + cols[4:-1]] # this line gives above error
# because cols[-1] returns the string 'Total' but we need an list to concatenate here
# Solution is to wrap it in list
```

```python
# This method basically does rearranging by concatenation
# And we are overwriting the value of original variable with the concatenation result
df = df[cols[0:4] + [cols[-1]] + cols[4:-1]]  # this line gives above error
df

# Using this method might not be robust, as it depends on indexing, and if indexes changes due to any reason,
# The final result of data might be messed up.
```

```python
# Write the data to a csv file
df.to_csv('../../data/modified_pokeman_data.csv', index=False)  # index=False will not save the index
# df.to_excel('../../../../data/modified_pokeman_data.csv', index=False)
df.to_csv('../../data/modified_pokeman_data.txt', index=False, sep='\t')
```

## Filtering Data

```python
# And &
filtered_data = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
filtered_data
```

```python
# OR |
filtered_data = df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
filtered_data
```

```python
# To reset index on filtered data,
# filtered_data = filtered_data.reset_index(drop=True) # drop=True will drop the old index
filtered_data.reset_index(drop=True,
                          inplace=True)  # drop=True, inplace=True will replace the original variable with new indices, so no need to reassign to same variable to overwrite
filtered_data
```

```python
# in python ~ represents not operator
# filtering Names containing Mega
df.loc[df['Name'].str.contains('Mega')]
```

```python
# filtering Names not containing Mega
df.loc[~df['Name'].str.contains('Mega')]
```

```python
df.loc[df['Type 1'].str.contains('fire|grass', regex=True)]  # Case issue, grass|fire first letters should be capital
```

```python
# library contains regex flags
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]  # makes the regex flag to make it case insensitive
```

```python
# filtering name with regex
df.loc[df['Name'].str.contains('^pi.*', flags=re.I, regex=True)]  # makes the regex flag to make it case insensitive
```

## Conditional Changes

```python
# Make a copy of the data frame
test_data = df.copy()

# Changes Legendary Value to True based on the Condition Type 1 == Fire
test_data.loc[test_data['Type 1'] == 'Fire', 'Legendary'] = True
test_data
```

```python
test_data.loc[(test_data['Type 1'] == 'Fire') & (test_data['Legendary'] == False)]
```

```python
# a test copy of the data
dft = df.copy()

# Changing multiple values based on condition
dft.loc[dft['Total'] > 500, ['Generation', 'Legendary']] = 'Test Value'
# Changes both properties values to "Test Value"

dft
```

```python
# Changing multiple values based on condition
dft.loc[dft['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
# Changes Generation to "Test 1" and Legendary to "Test 2"

dft
```

## Aggregate Statistics (Groupby)

```python
dft = df.copy()

# dft.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)

dft.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
# type(dft.groupby(['Type 1']))
dft.groupby(['Type 1']).count()['count']
```

## Working with large amounts of data
