# Line and bar plots in python


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
data = sns.load_dataset("iris")
data
```

```python
data.sample()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

## adding titles


```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Iris plot")
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Iris plot")
plt.xlim(1)
plt.ylim(1)
plt.show()
```

## Setting Styles
- darkgrid
- whitegrid
- dark
- white
- ticks


```python
sns.set_style('whitegrid')

sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Iris plot")
plt.show()
```

## Figure Size

```python
sns.set_style('whitegrid')
plt.figure(figsize=(16, 12))

sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Iris plot")
plt.show()
```

```python
sns.set_style('whitegrid')

sns.barplot(x="species", y="sepal_width", data=data)
plt.title("Iris plot")
plt.show()
```

```python
t_data = sns.load_dataset('titanic')
sns.barplot(x='who', y='alone', hue='sex', data=t_data)
# plt.xlabel('Who')
# plt.ylabel('Alone')
plt.show()
```

```python
t_data = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone', hue='who', data=t_data)
plt.xlabel('Gender')
plt.ylabel('Alone')
plt.show()
```

```python
t_data = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone', hue='who', data=t_data, order=["female", "male"])
plt.xlabel('Gender')
plt.ylabel('Alone')
plt.show()
```

```python
t_data = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone', hue='who', data=t_data, order=["female", "male"], color="red")
plt.xlabel('Gender')
plt.ylabel('Alone')
plt.show()
```

```python
t_data = sns.load_dataset('titanic')
sns.barplot(x='sex', y='alone', hue='who', data=t_data, order=["female", "male"], color="red", ci=None)
plt.xlabel('Gender')
plt.ylabel('Alone')
plt.show()
```

```python
from numpy import mean
sns.barplot(x='sex', y='alone', hue='who', data=t_data, order=["female", "male"], color="red", ci=None, palette='pastel', estimator=mean, saturation=0.5)
plt.show()
```
