# Working with large amounts of data


```python
new_df = pd.DataFrame(columns=df.columns)

# data = original_data.copy()
# data.groupby(['Type 1']).count()
# df['Type 1']
# reading data in chunks and then combining it.
for df in pd.read_csv('../../../data/pokemon_modified_data.csv', chunksize=5):
    result = df.groupby(['Type 1']).count()
    print(result)
    new_df = pd.concat([new_df, result])
    break

new_df
```
