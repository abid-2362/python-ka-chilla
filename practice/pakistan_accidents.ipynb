```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/traffic_annual_accidents.csv', header=1)
data.columns
```

```python
data.rename(columns={'Unnamed: 0': 'location'}, inplace=True)
```

```python
plt.figure(figsize=(20, 12))
accidents_plot = sns.barplot(x='location', y='Total number of accidents', hue='Year', data=data, palette='pastel')
accidents_plot.set_xticklabels(accidents_plot.get_xticklabels(), rotation=0, ha="center")
```

```python
plt.figure(figsize=(20, 12))
accidents_plot = sns.barplot(x='Killed', y='Total number of accidents', hue='Year', data=data, palette='pastel')
accidents_plot.set_xticklabels(accidents_plot.get_xticklabels(), rotation=90, ha="center")
```

```python
data[['location']].value_counts()
```

```python
islamabad_accidents = data.loc[data['location'] == 'Islamabad']
islamabad_accidents.columns
```

```python
sns.barplot(x='Year', y='Total number of accidents', hue='Killed', data=islamabad_accidents)
plt.title('Islamabad Accidents 2012 to 2019')
```

```python
islamabad_accidents_info = islamabad_accidents['Killed'].describe()
islamabad_accidents_info
```

```python
islamabad_accidents_info.loc['total killed'] =  islamabad_accidents['Killed'].sum()
islamabad_accidents_info
```
